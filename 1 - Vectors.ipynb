# Vectors

```python
# Definition : A vector is a quantity with magnitude and direction .
# Representation
#Geometric: Arrow from the origin (0,0,‚Ä¶,0) to the point (x,y,‚Ä¶).
#Algebraic: Row or column of real numbers.
```

```python
import numpy as np
from math import inf
```

```python
v1=np.array([1,2,-3])
v2=np.array([2,4,5])
```

## Basic operations

```python
# Addition and Subtraction
addition = v1 + v2
subtraction = v1 - v2
```

```python
# Scalar Multiplication
v1_doubled = 2 * v1 
```

```python
# Norm : The norm of a vector measures its length (size) in space.
# Common Types of Norms : L1 Norm - L2 Norm  - Infinity Norm
# using python : norm = np.linalg.norm(v, norm_type "Default 2") 
L1_norm =  np.linalg.norm(v1,1)
L2_norm =  np.linalg.norm(v1,2)
Linf_norm =  np.linalg.norm(v1,inf)
```

```python
# Normalization : the process of scaling a vector so that its length (magnitude) becomes 1, while keeping the same direction.
v2_normalized = v2 / np.linalg.norm(v2)          #equals the vector divided by it's L2_norm
np.linalg.norm(v2_normalized)
```

```python
# Dot Product (Inner Product):The dot product of two vectors measures how much they point in the same direction.
product = v1 @ v2
```

```python
# widely used for comparing vectors and helps measure projection of one vector onto another
# projection :The projection of a vector (ùë¢) onto another vector (ùë£) is the part of (u) that lies in the direction of (ùë£) .
v =np.array([5,1])
u =np.array([2,4])
proj_u_on_v = (u@v/(np.linalg.norm(v))**2) * v
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.set_xlim(0, 5)
ax.set_ylim(0, 4)
ax.set_aspect('equal')
ax.grid(True)
# Function to draw vectors
def draw_vector(start, vec, color, label):
    ax.quiver(start[0], start[1], vec[0], vec[1],
              angles='xy', scale_units='xy', scale=1,
              color=color, label=label, linewidth=2)
    
# Draw u (blue) from origin
draw_vector([0,0], u, 'blue', 'u')
# Draw v (green) from origin
draw_vector([0,0], v, 'green', 'v')

# Draw projection (red) from origin
draw_vector([0,0], proj_u_on_v, 'red', 'Projection of u on v')

# Draw dashed line from u to projection point (to show perpendicular drop)
ax.plot([u[0], proj_u_on_v[0]], [u[1], proj_u_on_v[1]], 'k--', alpha=0.7)

# Settings
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.title("Vector Projection: u onto v")

plt.show()
```
